```python
a = 10
print(a) # imprime el valor de a, que es 10
a = 20
print(a) # imprime el nuevo valor de a, que es 20
b = 20
print("La suma de a más b es: " + str(a + b)) # imprime la suma de a más 20, que es 40, convirtiendo el resultado a cadena para concatenar con el texto
print("El valor de a es: " + str(a) + " y el valor de b es: " + str(b)) # imprime los valores actuales de a y b, que son 20 y 20 respectivamente, convirtiendo ambos a cadena para concatenar con el texto
print(f"El valor de a es: {a} y el valor de b es: {b}") # imprime los valores actuales de a y b, que son 20 y 20 respectivamente, utilizando f-string para insertar las variables directamente en la cadena
print("El valor de a es:", a, "y el valor de b es:", b) # imprime los valores actuales de a y b, que son 20 y 20 respectivamente, utilizando comas para separar los argumentos en la función print, lo que añade espacios automáticamente entre ellos
print(f"El valor de a es: {a} y el valor de b es: {b} y la suma es {a+b}") # imprime los valores actuales de a y b, que son 20 y 20 respectivamente, y también la suma de ambos, que es 40, utilizando f-string para insertar las variables directamente en la cadena
```

### Convención de nombres para las variables

```python
a = 10 # a no es nombre significativo, es mejor usar nombres descriptivos para las variables
sumando1 = 10 # nombre más descriptivo para la variable que representa el primer número a sumar
sumando2 = 20 # nombre más descriptivo para la variable que representa el segundo número a sumar
suma = sumando1 + sumando2 # nombre descriptivo para la variable que almacena el resultado de la suma
print(f"La suma de {sumando1} y {sumando2} es: {suma}") # imprime el resultado de la suma utilizando f-string para insertar las variables directamente en la cadena, con nombres descriptivos para mayor claridad
```

```python
primer_sumando = 10 # nombre descriptivo para la variable que representa el primer número a sumar. Snake case es una convención de nombrado común en Python, donde las palabras se separan por guiones bajos para mejorar la legibilidad.
segundo_sumando = 20 # nombre descriptivo para la variable que representa el segundo número a sumar, siguiendo la convención de snake case
resultado_suma = primer_sumando + segundo_sumando # nombre descriptivo para la variable que almacena el resultado de la suma, siguiendo la convención de snake case
print(f"La suma de {primer_sumando} y {segundo_sumando} es: {resultado_suma}") # imprime el resultado de la suma utilizando f-string para insertar las variables directamente en la cadena, con nombres descriptivos siguiendo la convención de snake case para mayor claridad
```

```python
primerSumando = 10 # nombre descriptivo para la variable que representa el primer número a sumar. Camel case es una convención de nombrado común en algunos lenguajes de programación, donde las palabras se escriben juntas y cada palabra comienza con mayúscula para mejorar la legibilidad.
segundoSumando = 20 # nombre descriptivo para la variable que representa el segundo número a sumar, siguiendo la convención de camel case
resultadoSuma = primerSumando + segundoSumando # nombre descriptivo para la variable que almacena el resultado de la suma, siguiendo la convención de camel case
print(f"La suma de {primerSumando} y {segundoSumando} es: {resultadoSuma}") # imprime el resultado de la suma utilizando f-string para insertar las variables directamente en la cadena, con nombres descriptivos siguiendo la convención de camel case para mayor claridad
```
